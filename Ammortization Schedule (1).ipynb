```python
import pandas as pd
# Getting user input, validating it, turning into float (removing $,% and making sure it is able to be turned into float)
while True:
    try:
        principal = input("What is your principle amount (in $)?:")
        principal = float(principal)
        if principal > 0:
            break
    except:
        characters_to_remove = '$%,' 
        for i in characters_to_remove:
            principal = principal.replace(i, "")
        principal = float(principal)
        if principal > 0:
            break
        
while True:
    try:
        rate = input("What is your interest rate (in %)?:")
        rate = (float(rate))/100
        if rate > 0:
            break
    except:
        characters_to_remove = '$%,' 
        for i in characters_to_remove:
            rate = rate.replace(i, "")        
        rate = (float(rate))/100
        if rate > 0:
            break
        

while True:
    try:
        pmt = input("What is your payment (in $)?:")
        pmt = float(pmt)
        if pmt > 0 and pmt < principal:
            break
    except:
        characters_to_remove = '$%,' 
        for i in characters_to_remove:
            pmt = pmt.replace(i, "")
        pmt = float(pmt)
        if pmt > 0 and pmt < principal:
            break
#creating empty lists to append values to
month = 1
Month = []
Payment = []
Interest = []
Balance = []

# while loop in order to see how many payments are needed, and the interest portion and ending balance for each payment, then append each value to respective lists
while principal > pmt:
    interest = ((1/12)*rate)*principal
    principal = principal - (pmt - interest)
    Month.append(month)
    Payment.append(pmt)
    Interest.append(interest)
    Balance.append(principal)
    month += 1

# last payment will not be in while loop because payment will be equal to balance - finding month, payment, interest, and final balance (0) which will be appended to lists
Month.append(month)
interest = ((1/12)*rate)*principal
pmt = principal
Payment.append(pmt)
Interest.append(interest)
principal = principal - pmt
Balance.append(principal)

#formatting items in list for print statment 
sum1 = sum(Payment)
Payment = ['${:.2f}'.format(item) for item in Payment]
Balance = ['${:.2f}'.format(item) for item in Balance]
Interest = ['${:.2f}'.format(item) for item in Interest]
years = len(Month)//12
months = len(Month)%12

#creating output 
df = pd.DataFrame(list(zip(Month,Payment,Interest,Balance)), columns = ['Month','Payment','Interest','Balance'])
print(df)
print('You paid a total of ${:,.2f} over {} years and {} months.'.format(sum1, years, months))
```
